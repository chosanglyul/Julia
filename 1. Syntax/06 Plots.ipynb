```julia
using Pkg
Pkg.add("Plots")
using Plots
```

```julia
x = -3:0.1:3
f(x) = x^2
y = f.(x)
```

```julia
gr()
```

```julia
plot(x, y, label="line")
scatter!(x, y, label="points")
```

```julia
Pkg.add("PlotlyJS")
plotlyjs()
```

```julia
plot(x, y, label="line")
scatter!(x, y, label="points")
```

```julia
temp = [14.4, 14.5, 16.5]
num = [40000, 20000, 16660]
```

```julia
plot(temp, num, legend=false)
scatter!(temp, num, legend=false)
xflip!()
xlabel!("X")
ylabel!("Y")
title!("T")
```
